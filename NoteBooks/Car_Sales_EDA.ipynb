```python
cs['Owner_Type']=cs['Owner'].str.extract(r'(\d+)').astype(int)   #Removing st,rd from 'Owner' column
```

```python
cs[['Owner_Type','Price']].corr()
```

```python
cs.groupby('Brand')['Owner_Type'].value_counts()
```

```python
cs[['Brand','Owner_Type']].value_counts().reset_index(name='Count')
```

```python
# Visualizing on whether price decreases if owner type increases
sns.regplot(data=cs, x='Owner_Type', y='Price', scatter_kws={'s': 10, 'alpha': 0.5}, line_kws={'color': 'green'})
plt.xlabel('Owner Type')
plt.ylabel('Price')
plt.title('Price vs Owner type comparison')
plt.show()
```

```python
# Visualizing on whether price decreases if km increases
sns.regplot(data=cs, x='Kilometers', y='Price', scatter_kws={'s': 10, 'alpha': 0.3}, line_kws={'color': 'red'})
plt.xlabel('Kilometers')
plt.ylabel('Price')
plt.title('Price vs Kilometers comparison')
plt.show()
```
